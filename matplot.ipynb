```python
import matplotlib.pyplot as plt
```

## Grafik pada python

```python
x = [2,4]
y = [4,6]
plt.plot(x, y)
plt.title('Grafik Sederhana')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.show()
```

### Custom judul dan label

```python
plt.plot(x, y,
         color='red',
         label='y = 2x')

plt.title('Grafik Sederhana',
          fontfamily='Fira Code')

plt.xlabel('Sumbu X', 
           fontsize=10,
           fontweight='bold',
           color='blue',
           labelpad=10,
           loc='center',
           backgroundcolor='lightgray')

plt.ylabel('Sumbu Y',
           fontsize=10)

plt.grid(True,
         alpha=0.3)

plt.legend(loc='best',
           title='Data',
           facecolor='lightgray',
           edgecolor='black',
           shadow=True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(-3,3.1,0.1)
y1 =  x * 2
y2 = pow(x,2)
y3 = pow(x,3)

plt.plot(x,y1,
         color = 'orange',
         label = 'Y = 2x')

plt.plot(x, y2,
         color = 'green',
         label = 'Y = x ^ 2')

plt.plot(x,y3,
         color = 'blue',
         label = 'Y = x ^ 3')

plt.xlabel('x = {-3,3 | x = R}')

plt.grid(True)

plt.legend(loc='upper center',
           title='Keterangan',
           facecolor='lightgray',
           edgecolor='black',
           shadow=True,
           ncol = 3)

plt.show()
```
